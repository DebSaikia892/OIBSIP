# App Market Exploration and Analysis

```python
# App Market Exploration and Analysis

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
apps_data = pd.read_csv('apps.csv')
```

```python
# Data Cleaning
apps_data['Installs'] = apps_data['Installs'].str.replace(',', '').str.replace('+', '').astype(float)
apps_data['Price'] = apps_data['Price'].str.replace('$', '').astype(float)
```

```python
# Category Exploration

# Distribution of apps among categories
plt.figure(figsize=(10, 8))
sns.countplot(data=apps_data, x='Category')
plt.xticks(rotation=90)
plt.title('Apps Distribution Across Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Apps')
plt.show()
```

```python
# Number of unique categories
unique_categories = apps_data['Category'].nunique()
print(f'Total unique categories: {unique_categories}')
```

```python
# Average rating per category
print('Average Rating per Category:')
print(apps_data.groupby('Category')['Rating'].mean().sort_values())
```

```python
# Distribution of app type within each category
plt.figure(figsize=(10, 8))
sns.countplot(data=apps_data, x='Category', hue='Type')
plt.xticks(rotation=90)
plt.title('App Type Distribution Across Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Apps')
plt.legend(title='App Type')
plt.show()
```

```python
# Average number of reviews for apps in each category
avg_reviews_by_category = apps_data.groupby('Category')['Reviews'].mean().sort_values(ascending=False)
print('Average Reviews per Category:')
print(avg_reviews_by_category)
```

```python
# Top 5 categories with the highest number of installs
top_categories_installs = apps_data.groupby('Category')['Installs'].sum().nlargest(5)
print('Top 5 Categories with Highest Installs:')
print(top_categories_installs)
```

```python
# Distribution of app prices within each category
plt.figure(figsize=(10, 8))
sns.barplot(data=apps_data, x='Category', y='Price')
plt.xticks(rotation=90)
plt.title('Distribution of App Prices Across Categories')
plt.xlabel('Categories')
plt.ylabel('App Price ($)')
plt.show()
```

```python
# Metrics Analysis

# Rating Analysis
rating_stats = apps_data['Rating'].describe()
print('Rating Statistical Analysis:')
print(rating_stats)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(apps_data['Rating'], bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Size Analysis
size_stats = apps_data['Size'].describe()
print('Size Statistical Analysis:')
print(size_stats)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(apps_data['Size'], bins=20, kde=True)
plt.title('Distribution of App Sizes')
plt.xlabel('App Size')
plt.ylabel('Frequency')
plt.show()
```

```python
# Popularity Analysis (Installs)
install_stats = apps_data['Installs'].describe()
print('Installs Statistical Analysis:')
print(install_stats)
```

```python
plt.figure(figsize=(10, 8))
avg_installs_by_category = apps_data.groupby('Category')['Installs'].mean().sort_values(ascending=False)
avg_installs_by_category.plot(kind='bar')
plt.title('Average Installs Across Categories')
plt.xlabel('Categories')
plt.ylabel('Average Installs')
plt.xticks(rotation=90)
plt.show()
```

```python
# Price Analysis
price_stats = apps_data['Price'].describe()
print('Price Statistical Analysis:')
print(price_stats)
```

```python
plt.figure(figsize=(10, 8))
avg_prices_by_category = apps_data.groupby('Category')['Price'].mean().sort_values(ascending=False)
avg_prices_by_category.plot(kind='bar')
plt.title('Average Prices Across Categories')
plt.xlabel('Categories')
plt.ylabel('Average Prices')
plt.xticks(rotation=90)
plt.show()
```

```python
# Trend Analysis
apps_data['Last Updated'] = pd.to_datetime(apps_data['Last Updated'])
ratings_over_time = apps_data.groupby('Last Updated')['Rating'].mean()
```

```python
plt.figure(figsize=(10, 8))
ratings_over_time.plot()
plt.title('Average App Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Sentiment Analysis

user_reviews = pd.read_csv('user_reviews.csv')
```

```python
# Merge dataframes
merged_df = pd.merge(apps_data, user_reviews, on='App').dropna(subset=['Sentiment', 'Translated_Review'])
```

```python
# Sentiment Polarity Distribution
plt.figure(figsize=(11, 8))
sns.boxplot(x='Sentiment_Polarity', y='Type', data=merged_df)
plt.title('Distribution of Sentiment Polarity')
plt.xlabel('Sentiment Polarity')
plt.ylabel('App Type')
plt.show()
```

```python
plt.figure(figsize=(11, 8))
sns.barplot(x='Category', y='Sentiment_Polarity', data=merged_df)
plt.title('Sentiment Polarity Across Categories')
plt.xlabel('Categories')
plt.ylabel('Sentiment Polarity')
plt.xticks(rotation=90)
plt.show()
```
